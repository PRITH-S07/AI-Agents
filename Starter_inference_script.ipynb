```python
import os
import zipfile
import fitz  # PyMuPDF
import shutil

def extract_relevant_text(pdf_path):
    doc = fitz.open(pdf_path)
    full_text = ""

    for page in doc:
        text = page.get_text()
        if "Academic Integrity" in text:
            truncated_text = text.split("Academic Integrity")[0]
            full_text += truncated_text
            break
        else:
            full_text += text + "\n"

    doc.close()
    return full_text.strip()

def process_single_zip(zip_file_path, output_dir="cleaned_data"):
    os.makedirs(output_dir, exist_ok=True)
    temp_dir = "temp_pdf_extraction"

    # Extract the zip file
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(temp_dir)

    # Walk through extracted PDFs
    for root, _, files in os.walk(temp_dir):
        for file in files:
            if file.startswith("._") or not file.endswith(".pdf"):
                continue  # skip AppleDouble files and non-PDFs

            pdf_path = os.path.join(root, file)
            try:
                text = extract_relevant_text(pdf_path)
                if text:  # only save non-empty extractions
                    txt_filename = os.path.splitext(file)[0] + ".txt"
                    txt_path = os.path.join(output_dir, txt_filename)
                    with open(txt_path, "w", encoding="utf-8") as f:
                        f.write(text)
                    print(f"✅ Processed: {file}")
                else:
                    print(f"⚠️ No extractable text in {file}")
            except Exception as e:
                print(f"❌ Failed to process {file}: {e}")

    # Clean up extracted files
    shutil.rmtree(temp_dir)
    print(f"\n✅ Done. All cleaned .txt files saved to: {output_dir}/")

# Example usage
process_single_zip("/content/syllabi_pdfs_fall2024ECE.zip")
```

```python
from llama_index.core import VectorStoreIndex, ServiceContext, Document
from llama_index.core.node_parser import SimpleNodeParser
import os

# Load .txt files into LlamaIndex Documents
def load_txt_documents(folder_path):
    docs = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            with open(os.path.join(folder_path, filename), "r", encoding="utf-8") as f:
                text = f.read()
                docs.append(Document(text=text, metadata={"filename": filename}))
    return docs

documents = load_txt_documents("/content/cleaned_texts")
```

```python
from llama_index.llms.huggingface import HuggingFaceLLM
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Replace with your model — quantized or base
model_name = "TheBloke/Mistral-7B-Instruct-v0.1-GPTQ"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype=torch.float16,
    trust_remote_code=True
)

# Now use the standard constructor — NOT from_model_id
llm = HuggingFaceLLM(
    model=model,
    tokenizer=tokenizer,
    context_window=4096,
    max_new_tokens=512,
    tokenizer_name=model_name,
    query_wrapper_prompt="### Instruction:\n{query_str}\n\n### Response:",
    generate_kwargs={"temperature": 0.7, "do_sample": True}
)
```

```python
from llama_index.core import Settings
from llama_index.core import VectorStoreIndex
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# Use local MiniLM as embedder
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Set global config
Settings.llm = llm
Settings.embed_model = embed_model
```
